```python
def distance(origin, other):
    return np.sum((origin - other) ** 2)**(1/2)
```

```python
from collections import defaultdict
import operator

class radius_data():
    def __init__(self, x, y):
        self.data = x
        self.label = y
    
    def get_data(self, index = None, ignore_classes = {}):
        ignored = 0
        if index:
            return (self.data[index], self.label[index])
        else:
            i = 0
            while i < len(self.data):
                if self.label[i] not in ignore_classes:
                    yield (self.data[i], self.label[i])
                else:
                    ignored += 1
                i += 1
#             print(len([(self.data[i], self.label[i]) for i in range(len(self.data))]))
#             print([b for b in self.label])
#             return [(self.data[i], self.label[i]) for i in range(len(self.data))]

#         print("ignored", ignored, "items in dataset")

    def num_classes(self):
        return len(set(self.label))
    
    def get_classes(self):
        return list(set(self.label))
    
    def get_class_number(self):
        return {i: list(self.label).count(i) for i in set(self.label)}

class radius_nn_class():
    def __init__(self, weights = '', radius = 1, outlier_label = -1):
        self.weights = weights # for future use if there are different weights applied to distances
        self.radius = radius
        self.outlier_label = -1
    
    def fit(self, data: radius_data):
        # Takes all data and fits/scores all the data
        self.data = data
        self._poission_average()
#         print("did average")
    
    def predict(self, data, exact = False):
        # predict class for a given data point
        classes = defaultdict(int)
        i = 0
#         print(self.data.data)
        for (point, label) in self.data.get_data():
            i += 1
#             print(distance(point,data))
            if distance(point, data) <= self.radius:
                classes[label] += 1
#                 print(classes)
#                 print(distance(point,data), label)
#                 print(classes)
#                 print("data point number:", i)
#                 print(point)
        for label in classes:
#             print(classes[label], self.poission_amt[label])
            classes[label] = sp.stats.poisson.cdf(classes[label], self.poission_amt[label][label])
        if exact:
            return classes
        
        if len(classes) > 0:
            return max(classes.items(), key=operator.itemgetter(1))[0]
        return self.outlier_label
    
    def score(self, new_data = ""):
        #percentage correct on training data
        total = 0
        correct = 0
        if new_data == "":
            for (point, label) in self.data.get_data():
                if self.predict(point) == label:
                    correct += 1
                total += 1
        else:
            for (point, label) in new_data.get_data():
                if self.predict(point) == label:
                    correct += 1
                total += 1
        
        return correct/total
    
    def set_radius(self, radius):
        #change the radius
        self.radius = radius
        
    def predict_proba(self, data, raw = False):
        #predicts probability using raw numbers
        predictions = self.predict(data, True)
        precentages = {}
        for key in predictions.keys():
            precentages[key] = predictions[key] if raw else predictions[key]/sum(predictions.values())
        return precentages
    
    def _poission_average(self):
        #gets the mu for poisson distribution
        avg_amount = {}
        count = 0
        prev_label = None
        old_label = set()
        for (point1, label1) in self.data.get_data():
            if prev_label != label1:
                old_label.add(prev_label)
                prev_label = label1
            for(point2, label2) in self.data.get_data(ignore_classes = old_label):
                if distance(point1, point2) <= self.radius:
                    if label1 != label2:
                        if label1 in avg_amount:
                            avg_amount[label1][label2] = avg_amount[label1][label2]+1 if label2 in avg_amount[label1] else 1
                        else:
                            avg_amount[label1] = {label2: 1}
                        if label2 in avg_amount:
                            avg_amount[label2][label1] = avg_amount[label2][label1] + 1 if label1 in avg_amount[label2] else 1
                        else:
                            avg_amount[label2] = {label1: 1}
                    else:
                        if label1 in avg_amount:
                            avg_amount[label1][label2] = avg_amount[label1][label2]+1 if label2 in avg_amount[label1] else 1
                        else:
                            avg_amount[label1] = {label2: 1}
#             old_label.add(label1)
#             count += 1
        total_number = self.data.get_class_number()
        for label, counts in avg_amount.items():
            for counted, count in counts.items():
                counts[counted] = count/(total_number[label])
        self.poission_amt = avg_amount
#         print("here")
        
    def poission_average(self):
        #just returns the average counts
        return self.poission_amt
    
    def get_radius(sef):
        return sef.radius
    
    def predict_prob_poission(self, data, raw = False):
        #given data point, give percentages using poisson distribution
        #returns dictionary of {label: probabilty of point being that far away from expected}
        # if raw, gives the precentages for each
        # if not raw takes softmax of percentages so that they add up to one
        # if not in any class, both raw and not raw will return 0 for all
        
        def softmax(items):
            total = np.sum([value for value in items.values()])
            for key in items.keys():
                items[key] /= total if total != 0 else 0 
            return items
        
        def extreme(x, mu):
            classes = self.data.get_classes()
#             print(classes)
            xi = [x[classes[i]] if classes[i] in x else 0 for i in range(len(classes))]
            mui = [mu[classes[i]] if classes[i] in mu else 0 for i in range(len(classes))]
#             print(xi)
#             print(mui)
            values = [0 if mui[i] != 0 else int(xi[i] == 0) for i in range(len(classes))]
            def individual_poisson(x, mu):
                prob = math.e**(-mu)
                for i in range(x):
                    prob *= mu
                    prob /= (i+1)
#                 print(prob)
                return prob
            
            for i in range(len(classes)):
                for j in range(xi[i]):
                    values[i] += individual_poisson(j+1, mui[i])
#                     print(values[i])
            
            pure =  np.product(values)
            return pure*2 if pure <= .5 else (1-pure)*2
        
        actual = self.predict_proba(data, True)
        probs = {}
#         print(actual)
#         for label, value in self.poission_amt.items():
#             probs[label] = extreme(actual, value)
        print(actual)    
        return actual if raw else softmax(actual)
```
